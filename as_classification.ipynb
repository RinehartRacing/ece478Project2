```python
import matplotlib.pyplot as plt
fopen = open("20150801.as2types.txt", encoding="utf-8")
transit = 0
multihomed = 0
stubs = 0
for line in fopen:
    if line[0] != "#":
        line_split = line.split("|")
        as_class = line_split[2].strip("\n")
        if as_class == "Transit/Access":
            transit += 1
        elif as_class == "Enterpise":
            stubs += 1
        elif as_class == "Content":
            multihomed += 1
        else:
            print(f"Error with line {as_class}")

print(f"Transit = {transit}")
print(f"Multi-homed = {multihomed}")
print(f"Stubs = {stubs}")
x_ticks = ["Transit", "Multi-Homed", "Stubs"]
y_data = [transit, multihomed, stubs]
plt.bar(x_ticks, y_data)
for i in range(len(x_ticks)):
    plt.text(i, y_data[i], y_data[i], ha="center")
plt.title("2015 Autonomous System Class Distribution")
plt.show()
```
